# Visualization using python

```python
# import library
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
# histogram of petal length
df['petal length'].plot.hist(bins=20,figsize=(10, 10))
plt.show()
```

```python
# boxplot of four attributes.
df.plot.box(grid=False,figsize=(10, 10)) 
plt.show()
```

```python
# scatter plot of four attributes.
from pandas.plotting import scatter_matrix

# assign red, blue, and green for three classes.
color = []
for c in df['class']:
    if c == 'Iris-setosa':
        color.append('r')
    elif c == 'Iris-virginica':
        color.append('b')
    else:
        color.append('g')
        
scatter_matrix(df,figsize=(10, 10),marker='o', c=color)
plt.show()
```

```python
# parallel coordinates plot of Iris dataset.
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(15,10)) # predefine size of figure.
parallel_coordinates(df,'class') # 'class' is the column name of three classes.
plt.show()
```
